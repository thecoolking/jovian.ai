## Function 1 - np.isfinite() 

this checks each element in the array for finiteness and NaN (invalid or missing number)

```python
#example 1  
arr1 = [[1, 2], 
        [3, 4.]]
np.isfinite(arr1)
```

## Function 2 - np.islogical_and() 

function calculates the result of AND operation of the corresponding elements in two arrays

```python
# Example 1 - working
arr1 = [[1, 2,8], 
        [3, 4.,0]]

arr2 = [[5, 6, 7], 
        [8, 9, 10]]

np.logical_and(arr1, arr2)
```

In the above example, the value in the 0th position of the 0th array of arr1 is multiplied with 0th position of the 0th array of arr2 (note that the examples have 2 arrays of 3 elements each). The resultant array is checked if each individual element is True or False (integer or 0)

```python
# Example 2 - working
arr1=[[-3,0.56,9],[0.0004,12012,-912]]
arr2=[[4,998,0],[9898,3266,0.0292]]
np.logical_and(arr1,arr2)
```

Explanation about example

```python
# Example 3 - breaking (to illustrate when it breaks)
arr1=[[-3,0.56,9],[0.0004,12012,-912]]
arr2=[[4,998],[9898,3266,0.0292]]
np.logical_and(arr1,arr2)
```

Obviously in the above example, the dimensions of the 2 arrays are not the same. The dimensions of both arrays must be same for this to work. As per the docs, broadcasting can be done for 1 of the 2 arrays in the event that the dimensions do not match.

This function can be used to check the boolean multiplication between 2 arguments

## Function 3 - gcd()

gcd() calculates the greatest common divisor between the given arguments

```python
# Example 1 - working
np.gcd(34,20)
```

34 can be divide by 1,2,17 and itself. 20 can be divided by 1,2,4,5,10 and itself. So 2 is the gcd.

```python
# Example 2 - working
np.gcd(np.array([23,121,545]),np.array([989,33,23]))
```

Explanation about example

```python
# Example 3 - breaking (to illustrate when it breaks)
np.gcd(np.array([23,121]),np.array([989,33,23]))
```

Again the arrays have different size and cannot be used together

Arrays when having different sizes need to be broadcasted.

## Function 5 - lcm()

lcm() calculates the least common multiple of the given arguments

```python
# Example 1 - working
np.lcm(200,45)
```

1800 is the least common multiple of 200 and 45

```python
# Example 2 - working
np.lcm(9,12)
```

36 is the lcm of 9 and 12

```python
# Example 3 - breaking (to illustrate when it breaks)
np.lcm(2.5,0.5)
```

lcm() function does not work for float arguments
